# Using BASpy to read in climate model data (CMIP5) into Xarray

Requirements:
- [BASpy](https://github.com/scotthosking/baspy) (for obtaining data)
- [Xarray](http://xarray.pydata.org) (for data processing)
- [Cartopy](https://scitools.org.uk/cartopy) (for mapping visualizations)

````python
import baspy as bp
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import xarray as xr

%matplotlib inline
````

## Using BASpy

### Define the CMIP5 data we want to work with

````python
catlg = bp.catalogue(dataset='cmip5', Experiment='historical',
                         Frequency='mon', Var='tas',
                         Model='HadGEM2-ES')
catlg.drop(columns=['Path','DataFiles'])
````

### Select one model run (one row)

````python
row = catlg.iloc[3]
row
````

## Using Xarray

Load dataset using BASpy:
```python
ds = bp.open_dataset(row)
```

Or use the example dataset included with BASpy:

````python
ds = bp.eg_Dataset()
ds
````

````python
da = ds.tas
da
````

## Plotting

````python
da.plot()
````

### Plot map for first time index

````python
da.isel(time=0).plot()
````

### Plot using a polarstereo map projection

````python
crs = ccrs.SouthPolarStereo(central_longitude=0.0)
ax = plt.subplot(projection=crs)
ax.set_extent([-180,180,-90,-60], ccrs.PlateCarree() )
ax.gridlines(ylocs=range(-90,-30,5))
da.isel(time=0).plot.contourf(ax=ax, transform=ccrs.PlateCarree(),
                                  cmap=plt.cm.Blues_r,
                                  extend='both')
ax.coastlines('110m', color='k')
````
